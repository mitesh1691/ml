```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('emails.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# for null column
null_counts = df.isnull().sum()
# Filter columns with more than one null value
columns_with_nulls = null_counts[null_counts >= 1]
columns_with_nulls
```

```python
x = df.iloc[:,1:3001]
x.head()
```

```python
y = df.iloc[:,-1]
y.head()
```

```python
# Splitting the dataset into training and test set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.2 ,random_state=12)
```

```python
x_test
```

```python
y_test
```

```python
# Feature Scaling  
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
x_test
```

```python
x_train
```

# K-Nearest Neighbor(KNN)

```python
#Fitting K-NN classifier to the training set  
from sklearn.neighbors import KNeighborsClassifier  
knn= KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2 )  
knn.fit(x_train, y_train) 
```

```python
y_pred_knn = knn.predict(x_test)
y_pred_knn
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test, y_pred_knn)
cm
```

```python
from sklearn.metrics import accuracy_score
```

```python
#accuracy score
accuracy_score(y_test, y_pred_knn)
```

```python
sns.heatmap(cm, annot=True, fmt='.2f', cmap='Blues') #annot determines display value or not. fmt determines the values as decimal or float.
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```

# Support Vector Machine(svm)

```python
from sklearn.svm import SVC # "Support vector classifier"  
svc = SVC(kernel='linear', random_state=0)  
svc.fit(x_train, y_train)
```

```python
y_pred_svc = svc.predict(x_test)
y_pred_svc
```

```python
cm = confusion_matrix(y_test, y_pred_svc)
cm
```

```python
#accuracy score
accuracy_score(y_test, y_pred_svc)
```

```python
# Confusion matrix graph using seaborn
sns.heatmap(cm, annot=True, fmt='.2f', cmap='Blues') #annot determines display value or not. fmt determines the values as decimal or float.
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```
